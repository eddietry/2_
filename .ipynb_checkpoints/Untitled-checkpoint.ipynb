```python
import random
import math
```

```python
def objective_function(vector):
    return sum([ x**2 for x in vector])

def random_vector(minmax):
    sol = [round(random.uniform(minmax[i][0],minmax[i][1]),2) for i in range(len(minmax))]
    return sol
```

```python
def take_step(minmax, current, step_size):
    # 根据current解进行更新，返回新解
    position = []
    for id_,x in enumerate(current):
        min_i = max(minmax[id_][0],x-step_size) 
        max_i = min(minmax[id_][1],x+step_size)
        value = round(random.uniform(min_i,max_i),2)
        position.append(value)
    return position
```
